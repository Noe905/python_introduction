```python
numero_uno = input("Introduce el primer número")
numero_dos = input("Introduce el segundo número")

suma = int(numero_uno) + int(numero_dos)
print("Suma:", suma)
resta = int(numero_uno) - int(numero_dos)
multi = int(numero_uno) * int(numero_dos)
division = int(numero_uno) / int(numero_dos)

print("Suma:", suma, "\nResta:", resta, "\nMultiplicación:", multi, "\nDivisión:", division)
```

```python
numero_1 = int(input("Dame un número"))
numero_2 = int(input("Dame otro número"))
operacion_1y2= input("Introduce el tipo de operación")
while operacion_1y2 not in ("-" , "+", "/", "*") :
    operacion_1y2= input("Introduce el tipo de operación")

if operacion_1y2 == "-" :
    print("El resultado es:", abs(numero_1 - numero_2))
elif operacion_1y2 == "+" :
    print("El resultado es:", abs(numero_1 + numero_2))
elif operacion_1y2 == "/" :
    print("El resultado es:", abs(numero_1 / numero_2))
elif operacion_1y2 == "*" :
    print("El resultado es:", abs(numero_1 * numero_2))

```
